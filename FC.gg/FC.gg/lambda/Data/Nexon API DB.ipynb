```python
import pandas as pd

player = pd.read_csv('player.csv')
pid = pd.read_csv('player_pid.csv')
```

```python
player
```

```python
pid
```

```python
#pid에서 sid 컬럼 추가
pid['sid'] = pid['id'].astype(str).str[:3]

pid['sid']
```

```python
pid['sid'] = pid['sid'].astype('int')
```

```python
player['seasonId'] = player['seasonId'].astype('int')
```

```python
#pid와 player 합치기
merged_data = pd.merge(pid, player, left_on='sid', right_on='seasonId', how='inner')

# 선택된 행 출력
print(merged_data)
```

```python
merged_data['spid'] = merged_data['id'].astype('str').str[3:]

merged_data['spid']
```

```python
merged_data = merged_data.astype('str')
```

```python
def remove_leading_zeros(value):
    start_index = 0
    for i, char in enumerate(value):
        if char != '0':
            start_index = i
            break
    return value[start_index:]


data =  merged_data['spid'].apply(remove_leading_zeros)

data
```

```python
merged_data['spid'] = data
merged_data.drop('seasonId' , axis =1, inplace=True)
merged_data
```

```python
url_pattern = "https://fco.dn.nexoncdn.co.kr/live/externalAssets/common/players/p{sid}.png"

merged_data['p_image'] = merged_data['spid'].apply(lambda x: url_pattern.format(sid=x))

print(merged_data['p_image'])
```

```python
action_url = 'https://fco.dn.nexoncdn.co.kr/live/externalAssets/common/playersAction/p{sid}.png'
merged_data['p_action_image'] = merged_data['id'].apply(lambda x: action_url.format(sid=x))

print(merged_data['p_action_image'])
```

```python
import requests
import numpy as np

# 모든 URL에 대해 반복하여 null 여부 확인하고 null 값인 경우 해당 값을 null로 대체
for i, url in enumerate(merged_data['p_image']):
    response = requests.get(url)
    if response.status_code != 200:
        merged_data.at[i, 'p_image'] = np.nan

print(merged_data)
```
